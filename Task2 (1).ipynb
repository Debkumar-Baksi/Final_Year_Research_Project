```python
import networkx as nx
import matplotlib.pyplot as plt


# Create an empty graph
G = nx.Graph()


# Add nodes
node_input=int(input("Enter the maximum number of nodes : "))
for i in range(1,node_input+1):
    G.add_node(i)
link_disjoint_path_input=int(input(f"Enter number of Link Disjoint Paths (max : {node_input - 1}) : "))
    

# Add edges (edges connect nodes)
counter=0
for x in range(1,node_input+1):
    for y in range(1,node_input+1):
        if (x==y):
            continue
        G.add_edge(x,y)
        counter+=1
        
        
# Draw Graph
nx.draw_networkx(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=8)
plt.figure(figsize=(6, 4))
plt.show()


print(f"Here are {counter} distinct pair of {link_disjoint_path_input} Link Disjoint Paths : ")
print("\n")


for x in range(1,node_input+1):
    for y in range(1,node_input+1):
        if (x==y):
            continue
        source=x
        target=y
        
        
        paths = list(nx.edge_disjoint_paths(G, source, target))
        limited_paths = paths[:link_disjoint_path_input]

        #Print the edge-disjoint paths
        print(f"2 link-disjoint paths from {source} to {target}:")
        for i, path in enumerate(limited_paths, start=1):
            print(f"Path {i}: {path}")
        
        print("\n")
        y+=1
    x+=1
```
